```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('/content/final_silver.csv')

# Plot
df["time"] = pd.to_datetime(df["time"])
plt.figure(figsize=(14, 6))
plt.plot(df["time"], df['sensor.sensor_temperatura_1_temperature'], label='Sensor 1')
plt.plot(df["time"], df['sensor.sensor_temperatura_2_temperature'], label='Sensor 2')
plt.plot(df["time"], df['sensor.sensor_temperatura_3_temperature'], label='Sensor 3')
plt.plot(df["time"], df['sensor.sensor_temperatura_4_temperature'], label='Sensor 4')
plt.plot(df["time"], df['temperature ℃'], label='Temperatura Real (℃)', linestyle='-.')


plt.xlabel('Time')
plt.ylabel('Temperaturas')
plt.title('Temperaturas en horario de calefacción')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

En la gráfica de arriba se muestra un histórico de las temperaturas captadas de los senesores del aula, junto al histórico proporcionado por el instituto.

Durante el primer mes no se ven registros de los sensores 3 y 4, esto es debido a que tardaron un mes en activar estos sensores.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el CSV
df = pd.read_csv("/content/final_silver.csv")  # Reemplaza con la ruta de tu archivo

df_variables_temperaturas = df[["time", "sensor.sensor_temperatura_1_temperature", "sensor.sensor_temperatura_2_temperature",
                               "sensor.sensor_temperatura_3_temperature", "sensor.sensor_temperatura_4_temperature", "temperature ℃"]]

# Convertir la columna 'time' a formato datetime
df_variables_temperaturas["time"] = pd.to_datetime(df_variables_temperaturas["time"])

# Definir el intervalo de fechas (MODIFICA AQUÍ)
fecha_inicio = "2025-01-17 00:00:00"  # Fecha de inicio
fecha_fin = "2025-01-19 23:59:59"  # Fecha de fin

# Filtrar datos en el rango de fechas
df_filtrado = df_variables_temperaturas[(df_variables_temperaturas["time"] >= fecha_inicio) & (df_variables_temperaturas["time"] <= fecha_fin)]


plt.figure(figsize=(12, 6))
plt.plot(df_filtrado["time"], df_filtrado["sensor.sensor_temperatura_1_temperature"], linestyle="-", label="Sensor 1")
plt.plot(df_filtrado["time"], df_filtrado["sensor.sensor_temperatura_2_temperature"], linestyle="-", label="Sensor 2")
plt.plot(df_filtrado["time"], df_filtrado["sensor.sensor_temperatura_3_temperature"], linestyle="-", label="Sensor 3")
plt.plot(df_filtrado["time"], df_filtrado["sensor.sensor_temperatura_4_temperature"], linestyle="-", label="Sensor 4")
plt.plot(df_filtrado["time"], df_filtrado["temperature ℃"], linestyle="-", label="temperature ℃")

# Personalización del gráfico
plt.xlabel("Tiempo")
plt.ylabel("Temperatura (°C)")
plt.title("Comparación de Temperaturas entre Sensores (1-4)")
plt.xticks(rotation=45)  # Rotar etiquetas del eje X para mayor legibilidad
plt.legend(title="Sensores")
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

En la gráfica de arriba se muestra un histórico de las temperaturas  captadas de los senesores del aula desde el 17 de enero hasta el 20 de enero, ya que es en este lapso de tiempo que pusieron los sensores 3 y 4.

En esta gráfica podemos ver la diferencia que hay entre el sensor 2 y el resto. También, podemos ver que entre los sensores 1, 3 y 4 no hay mucha diferencias.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv('/content/final_silver.csv')

selected_columns = ['sensor.sensor_temperatura_1_temperature', 'sensor.sensor_temperatura_2_temperature', 'sensor.sensor_temperatura_3_temperature', 'sensor.sensor_temperatura_4_temperature']  # Reemplaza con las que quieras

# Crear y mostrar la matriz de correlación
correlation_matrix = df[selected_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Matriz de Correlación (Temperaturas)")
plt.tight_layout()
plt.show()
```

En la gráfica de correlación de arriba se muestra la relación de los sensores entre ellos. En este caso está relacionando las variables de temperatura.

En esta gráfica podemos ver que el sensor 2 es el que tiene peores correlaciones, esto es debido al lugar en el que se encuentra. Por el resto los 4 sensores son muy parecidos.

En la gráfica de correlación de arriba se muestra la relación de todas las variables del archivo csv entre ellas.

En esta gráfica podemos ver que la variable más relacionada con temperatura real es el sensor 1 de la temperatura.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos
df = pd.read_csv('/content/entrenamiento_neuronal_calefaccion_gold.csv')  # o pd.read_excel('tus_datos.xlsx')

# Seleccionar solo columnas numéricas
df = df.select_dtypes(include='number')

# Variable específica
variable_objetivo = 'temperatura_calefaccion_y'

# Calcular correlación de esa variable con las demás
correlaciones = df.corr()[[variable_objetivo]].drop(index=variable_objetivo)

# Crear heatmap (una sola columna)
plt.figure(figsize=(6, len(correlaciones) * 0.5))
sns.heatmap(correlaciones, annot=True, cmap='coolwarm', vmin=-1, vmax=1, cbar=True)
plt.title(f"Correlación de '{variable_objetivo}' con otras variables")
plt.tight_layout()
plt.show()
```

En la gráfica de correlación de arriba se muestra la relación de todas las variables del archivo csv con la temperatura_calefaccion_y (Target).

Esta gráfica es una forma más reducida de la gráfica anterior, en este se ve más claro las relaciones de las distintas variable con el target.

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('/content/predicciones_se_abrira.csv')

df = df.sort_values(by='time')

# Plot
df["time"] = pd.to_datetime(df["time"])
plt.figure(figsize=(14, 6))
plt.plot(df["time"], df['temperatura_calefaccion_y'], label='Temperatura Real')
plt.plot(df["time"], df['temperatura_predicha'], label='Temperatura Predicha')


plt.xlabel('Time')
plt.ylabel('Temperaturas')
plt.title('Temperaturas en horario de calefacción')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
